3*t**3 - 2*t**2 + 5
position , speed , acceleration
 t = 0,1,2,3,4,5 plot each of them as a function of time

```python
from numpy import diff
from sympy import Symbol
import pandas as pd
import matplotlib.pyplot as plt
t = Symbol('t')
X = 3*t**3 - 2*t**2 + 5
Xprime = X.diff(t)
Xdprime = X.diff(t,2)
v,a,x = [],[],[]
for i in range(0,6):
    x.append(X.subs(t,i))
    v.append(Xprime.subs(t,i))
    a.append(Xdprime.subs(t,i))
df = pd.DataFrame({"t":range(6),"x":x,"v":v,"a":a})
df
```

```python
plt.plot(df["t"],x)
plt.xlabel("t")
plt.ylabel("x")
plt.grid(True,linestyle = ":")
plt.show()
plt.plot(df["t"],v)
plt.xlabel("t")
plt.ylabel("v")
plt.grid(True,linestyle = ":")
plt.show()
plt.plot(df["t"],a)
plt.xlabel("t")
plt.ylabel("a")
plt.grid(True,linestyle = ":")
plt.show()
```
